```python
import pandas as pd

# Now you can read the Excel file
df = pd.read_excel('netflix_titles.xlsx')

# Optional: See the column names
df.columns
```

```python
#converting column into a list
movie_list = df['title'].tolist()

print(movie_list[:5])    # View first 5 items
print(len(movie_list))   # Total number of items
```

```python
from imdb import IMDb

# Create IMDb object
ia = IMDb()
```

```python
import time


# Assuming you've already uploaded and read your Excel file
# df = pd.read_excel('your_file.xlsx')
movie_list = df['title'].dropna().tolist()
```

```python
movie_data = []

for i, title in enumerate(movie_list):
    try:
        print(f"{i+1}/{len(movie_list)} - Searching: {title}")

        results = ia.search_movie(title)
        if results:
            movie = results[0]
            ia.update(movie)

            # Get director(s)
            directors = movie.get('directors', [])
            director_names = ', '.join([person['name'] for person in directors]) if directors else 'Not found'

            # Get cast (first 5 actors)
            cast = movie.get('cast', [])
            cast_names = ', '.join([person['name'] for person in cast[:5]]) if cast else 'Not found'

            # Get country
            countries = movie.get('countries', [])
            country_list = ', '.join(countries) if countries else 'Not found'
        else:
            director_names = cast_names = country_list = 'Not found'

        movie_data.append({
            'Movie': title,
            'Director': director_names,
            'Cast': cast_names,
            'Country': country_list
        })

    except Exception as e:
        movie_data.append({
            'Movie': title,
            'Director': f"Error: {str(e)}",
            'Cast': f"Error: {str(e)}",
            'Country': f"Error: {str(e)}"
        })

    time.sleep(0.3)  # Pause to avoid overloading IMDb
```

```python
# Convert movie_data to DataFrame
df_final = pd.DataFrame(movie_data)

# Save to Excel
df_final.to_excel('movie_director_cast_country_4000.xlsx', index=False)
```
